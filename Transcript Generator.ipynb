```python
from pydub import AudioSegment
import os
import numpy as np
import winsound
import csv
```

```python
def crop_audio_sample(base_file,start,finish):
    sample = base_file[start:finish]
    sample.export('temp_sample.wav', format="wav")
    return sample
```

```python
def prompt_transcript():
    
    accept_transcript = input('Transcript this audio? Y/N')
    
    if accept_transcript == 'Y' or accept_transcript == 'y':
        name = input('Who is this voice from? 1 Silvio 2 Bolsonaro')
        
        if name == str(1):
            name = 'Silvio'
        else:
            name = 'Bolsonaro'
        
        transcript = input('Transcript the audio:')
        transcript_flag = True
        return transcript, name
    else:
        return None, None
```

```python
def play_sample():
    winsound.PlaySound('temp_sample.wav', winsound.SND_FILENAME)
```

```python
def save_sample(sample,filepath,filename):
    full_path = os.path.join(filepath,filename)
    print(filepath)
    print(filename)
    sample.export(full_path, format="wav")
```

```python
def append_transcript(filename, transcript, transcript_path):
    row = f'{filename[:-4]}|{transcript}'
    with open(transcript_path,'a') as transcript_file:
        writer = csv.writer(transcript_file)
        writer.writerow([row])
```

```python
#CONSTANTS
initial_time = ((2*60)+35)*1000 #02:35
custom_datasets_path = r"E:\Projects\TextToSpeech\custom_datasets"
silvio_path = r"E:\Projects\TextToSpeech\custom_datasets\Silvio"
bolsonaro_path = r"E:\Projects\TextToSpeech\custom_datasets\Bolsonaro"
base_file = AudioSegment.from_wav("main_audio_silvio.wav")
cropped_base_file = base_file[initial_time:]
```

```python
start = 0
end = 0
silvio_counter = 0
bolsonaro_counter = 0

while start < len(cropped_base_file):
    duration = np.random.randint(2,5) * 1000
    end += duration
    
    sample = crop_audio_sample(cropped_base_file, start, end)
    play_sample()
    transcript, name = prompt_transcript()
    
    if transcript != None:
        if name == 'Silvio':
            silvio_counter += 1
            filepath = f'{silvio_path}/wavs'
            filename = f'{name}-{str(silvio_counter).zfill(4)}.wav'
            transcript_path = os.path.join(silvio_path,'transcript.csv')
        else:
            bolsonaro_counter += 1
            filepath = f'{bolsonaro_path}/wavs'
            filename = f'{name}-{str(bolsonaro_counter).zfill(4)}.wav'
            transcript_path = os.path.join(bolsonaro_path,'transcript.csv')
        
        
        save_sample(sample,filepath,filename)
        append_transcript(filename, transcript, transcript_path)
        
    start = end
```
