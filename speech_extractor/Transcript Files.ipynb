```python
from pydub import AudioSegment
import os
import numpy as np
import winsound
import csv
```

```python
def append_transcript(filename, transcript, transcript_path):
    row = f'{filename[:-4]}|{transcript}'
    with open(transcript_path,'a') as transcript_file:
        writer = csv.writer(transcript_file)
        writer.writerow([row])
```

```python
def save_sample(sample,filepath):    
    sample.export(filepath, format="wav")
```

```python
directory = './/cropped'
transcript_path = 'transcript.csv'
temp_sample_path = 'temp_sample.wav'
```

```python
for filename in os.listdir(directory):
    file_path = os.path.join(directory,filename)
    sample = AudioSegment.from_wav(file_path)
    
    save_sample(sample,temp_sample_path)
    
    transcript = input(f'Type the audio for {filename}\n')
    
    append_transcript(filename,transcript,transcript_path)
    
    #print(sample.duration_seconds)
```
